```python
import pandas
from pandas import read_csv
import numpy as np
import matplotlib.pyplot as plt

dataset = read_csv('PT_2020-03-29-resume.csv', sep=',', header=0)
dataset=dataset.drop(dataset.columns[0], axis=1)

names=dataset.plot_name.unique()

dfs= pandas.DataFrame(dataset.drop(columns=[dataset.columns[0]])).values

dfs= np.reshape(dfs,(6,43,2)) # 6 time series with 43 steps each with "value" and "date"

plt.rcParams["figure.figsize"] = (15,5)
for x in range(names.size):
    plt.plot(dfs[x,:,0],dfs[x,:,1])
    plt.ylabel(names[x])
    plt.gcf().autofmt_xdate()
   
    plt.show()
```
